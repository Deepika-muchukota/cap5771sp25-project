```python
# 3. Feature Engineering 
print("\n\n== Feature Engineering ==")

# Standardize column names in dealing_anxiety
dealing_anxiety_df.columns = [col.lower().replace('/', '_').replace(' ', '_') 
                             for col in dealing_anxiety_df.columns]

# Filter datasets for the most recent year data
filled_form_df_2020 = filled_form_df[filled_form_df['Year'] == 2020]
dealing_anxiety_df_2020 = dealing_anxiety_df[dealing_anxiety_df['year'] == 2020]  # Use lowercase 'year'

# Find common entities across datasets
common_entities = set(filled_form_df_2020['Entity']).intersection(
    set(dealing_anxiety_df_2020['entity'])  # Use lowercase 'entity'
)

print(f"Number of common entities across datasets: {len(common_entities)}")
print("Sample entities:", list(common_entities)[:10])

# Create features for each entity
features_list = []

for entity in common_entities:
    # Base feature dictionary
    feature_dict = {'entity': entity}
    
    # Add anxiety prevalence from filled_form dataset
    filled_form_row = filled_form_df_2020[filled_form_df_2020['Entity'] == entity]
    if not filled_form_row.empty:
        feature_dict['anxiety_prevalence'] = filled_form_row[
            'Share - Question: mh7a - Have been anxious/depressed - Answer: Yes - Gender: all - Age_group: all'
        ].values[0]
    
    # Add coping strategy data from dealing_anxiety dataset
    dealing_row = dealing_anxiety_df_2020[dealing_anxiety_df_2020['entity'] == entity]  # Use lowercase 'entity'
    if not dealing_row.empty:
        # Extract coping strategies (using lowercase column names)
        strategies = [
            "religious_spiritual_activities", 
            "improved_lifestyle", 
            "changed_work_situation", 
            "changed_relationships", 
            "talked_to_friends_family", 
            "took_medication", 
            "spent_time_outdoors", 
            "talked_to_professional"
        ]
        
        # Add each strategy value to features
        for strategy in strategies:
            if strategy in dealing_row.columns:
                feature_dict[strategy] = dealing_row[strategy].values[0]
        
        # Calculate derived features
        # Social support score
        if 'talked_to_friends_family' in feature_dict and 'changed_relationships' in feature_dict:
            feature_dict['social_support_score'] = (
                feature_dict['talked_to_friends_family'] + feature_dict['changed_relationships']
            ) / 2
        
        # Self-care score
        if 'improved_lifestyle' in feature_dict and 'spent_time_outdoors' in feature_dict:
            feature_dict['self_care_score'] = (
                feature_dict['improved_lifestyle'] + feature_dict['spent_time_outdoors']
            ) / 2
        
        # Determine dominant coping approach
        if all(strategy in feature_dict for strategy in strategies):
            category_scores = {
                'clinical': feature_dict['took_medication'] + feature_dict['talked_to_professional'],
                'lifestyle': feature_dict['improved_lifestyle'] + feature_dict['spent_time_outdoors'],
                'social': feature_dict['talked_to_friends_family'] + feature_dict['changed_relationships'],
                'work': feature_dict['changed_work_situation'],
                'spiritual': feature_dict['religious_spiritual_activities']
            }
            
            feature_dict['dominant_coping_approach'] = max(
                category_scores.items(), key=lambda x: x[1]
            )[0]
    
    features_list.append(feature_dict)

# Convert to DataFrame
features_df = pd.DataFrame(features_list)

# Filter out entities with incomplete data
complete_features_df = features_df.dropna(subset=[
    'anxiety_prevalence', 
    'social_support_score', 
    'self_care_score'
])

print(f"\nEngineered features for {len(complete_features_df)} entities with complete data")
print("\nSample of engineered features:")
display(complete_features_df.head())
```

```python
# 5. Define Intents and Response Templates
print("\n\n== Defining Chatbot Intents and Response Templates ==")

# Define the intents for the chatbot
intent_mapping = [
    {
        "intent": "anxiety_prevalence",
        "keywords": ["prevalence", "common", "how many", "percentage", "rate", "statistics"],
        "required_features": ["anxiety_prevalence"],
        "response_template": "In {entity}, approximately {anxiety_prevalence:.1f}% of people reported experiencing anxiety or depression in 2020."
    },
    {
        "intent": "medication_info",
        "keywords": ["medication", "meds", "drugs", "prescriptions", "pills"],
        "required_features": ["took_medication"],
        "response_template": "About {took_medication:.1f}% of people in {entity} use medication to manage anxiety or depression."
    },
    {
        "intent": "therapy_info",
        "keywords": ["therapy", "counseling", "professional", "psychologist", "psychiatrist", "therapist"],
        "required_features": ["talked_to_professional"],
        "response_template": "In {entity}, approximately {talked_to_professional:.1f}% of people dealing with anxiety or depression seek help from mental health professionals."
    },
    {
        "intent": "coping_strategies",
        "keywords": ["cope", "deal", "manage", "handling", "strategies", "methods", "help"],
        "required_features": ["dominant_coping_approach", "social_support_score", "self_care_score"],
        "response_template": "People in {entity} primarily rely on {dominant_coping_approach} approaches to cope with anxiety. {self_care_score:.1f}% focus on self-care activities like lifestyle improvements and spending time outdoors. {social_support_score:.1f}% turn to social support networks."
    },
    {
        "intent": "social_support",
        "keywords": ["friends", "family", "talk", "social", "support", "relationships"],
        "required_features": ["talked_to_friends_family", "social_support_score"],
        "response_template": "In {entity}, {talked_to_friends_family:.1f}% of people talk to friends and family when dealing with anxiety or depression. Social support is a key coping mechanism with an overall score of {social_support_score:.1f}%."
    },
    {
        "intent": "lifestyle_changes",
        "keywords": ["lifestyle", "exercise", "diet", "outdoors", "nature", "physical", "habits"],
        "required_features": ["improved_lifestyle", "spent_time_outdoors", "self_care_score"],
        "response_template": "In {entity}, {improved_lifestyle:.1f}% of people improve their lifestyle and {spent_time_outdoors:.1f}% spend time outdoors to manage anxiety and depression. The overall self-care score is {self_care_score:.1f}%."
    }
]

print(f"Defined {len(intent_mapping)} intents for the chatbot")
print("\nExample intent configuration:")
display(intent_mapping[0])
```

```python
# 6. Build Intent Classification Model
print("\n\n== Building Intent Classification Model ==")

# Create a corpus of example questions for TF-IDF training
intent_corpus = []
intent_labels = []

# Generate training examples for each intent
for intent_data in intent_mapping:
    # Create examples from keywords
    for keyword in intent_data["keywords"]:
        # Generate multiple templates for each keyword
        intent_corpus.append(f"What is the {keyword} of anxiety?")
        intent_labels.append(intent_data["intent"])
        
        intent_corpus.append(f"Tell me about {keyword} for anxiety")
        intent_labels.append(intent_data["intent"])
        
        intent_corpus.append(f"How many people {keyword} with anxiety?")
        intent_labels.append(intent_data["intent"])
        
        intent_corpus.append(f"What's the {keyword} in different countries?")
        intent_labels.append(intent_data["intent"])

print(f"Generated {len(intent_corpus)} training examples for intent classification")
print("Sample training examples:")
for i in range(min(5, len(intent_corpus))):
    print(f"- \"{intent_corpus[i]}\" -> {intent_labels[i]}")

# Initialize and fit TF-IDF vectorizer
tfidf_vectorizer = TfidfVectorizer(
    ngram_range=(1, 2),
    min_df=1,
    max_df=0.9
)

intent_vectors = tfidf_vectorizer.fit_transform(intent_corpus)

# Create a function to classify intents using TF-IDF similarity
def classify_intent(question):
    """
    Classify the intent of a user question using TF-IDF similarity.
    
    Args:
        question: The user's question as a string
        
    Returns:
        The identified intent as a string or None if no match
    """
    # Convert the question to a TF-IDF vector
    question_vector = tfidf_vectorizer.transform([question])
    
    # Calculate similarity to all training examples
    similarities = cosine_similarity(question_vector, intent_vectors).flatten()
    
    # Get the best match
    best_match_idx = np.argmax(similarities)
    best_score = similarities[best_match_idx]
    
    # Return the intent if the score is above threshold
    if best_score > 0.1:  # Adjust threshold as needed
        return intent_labels[best_match_idx]
    else:
        return None
```

```python
# 7. Build Entity Extraction Function
print("\n\n== Building Entity Extraction Function ==")

def extract_entity(question, entity_list):
    """
    Extract the entity (country/region) from a user question.
    
    Args:
        question: The user's question as a string
        entity_list: List of valid entities to check for
        
    Returns:
        The identified entity as a string or None
    """
    # Convert to lowercase for case-insensitive matching
    question_lower = question.lower()
    
    # Try to match entities in the question
    for entity in entity_list:
        if entity.lower() in question_lower:
            return entity
    
    return None

# Test the entity extraction function with a sample question
sample_question = "What's the prevalence of anxiety in Australia?"
entity = extract_entity(sample_question, chatbot_features_df['entity'].tolist())
print(f"Sample question: \"{sample_question}\"")
print(f"Extracted entity: \"{entity}\"")

# Test with a question that doesn't contain an entity
sample_question_no_entity = "What are some common coping strategies for anxiety?"
entity = extract_entity(sample_question_no_entity, chatbot_features_df['entity'].tolist())
print(f"Sample question without entity: \"{sample_question_no_entity}\"")
print(f"Extracted entity: {entity}")
```

```python
# 8. Response Generation Function
print("\n\n== Building Response Generator ==")

import re  # Make sure to import this for regex pattern matching

def generate_response(intent, entity, features_df, intent_mapping):
    """
    Generate a response based on the identified intent and entity.
    
    Args:
        intent: The identified intent
        entity: The identified entity (country/region)
        features_df: DataFrame containing entity features
        intent_mapping: List of intent configurations with response templates
        
    Returns:
        A response string
    """
    # Get entity features
    entity_features = features_df[features_df['entity'] == entity]
    
    if entity_features.empty:
        return f"I don't have information about {entity} for this topic."
    
    # Get the intent configuration
    intent_config = next((i for i in intent_mapping if i["intent"] == intent), None)
    
    if not intent_config:
        return "I don't understand that question."
    
    # Check if we have all required features
    entity_data = entity_features.iloc[0].to_dict()
    missing_features = [f for f in intent_config["required_features"] 
                       if f not in entity_data or pd.isna(entity_data[f])]
    
    if missing_features:
        return f"I don't have complete information about {intent} for {entity}."
    
    # Generate response using template
    response = intent_config["response_template"]
    
    # Replace placeholders with actual values
    for key, value in entity_data.items():
        placeholder = "{" + key + "}"
        if placeholder in response:
            if isinstance(value, (int, float)):
                formatted_value = value
            else:
                formatted_value = value
            
            # Format with placeholders that include formatting
            placeholder_with_format = re.findall(r"{" + key + r":[^}]+}", response)
            if placeholder_with_format:
                for p in placeholder_with_format:
                    format_spec = p[p.find(":") + 1:-1]
                    if isinstance(value, (int, float)):
                        formatted_text = format(value, format_spec)
                        response = response.replace(p, formatted_text)
            elif placeholder in response:
                response = response.replace(placeholder, str(formatted_value))
    
    return response

# Test the response generation with sample intent and entity
sample_intent = "anxiety_prevalence"
sample_entity = "Australia"
response = generate_response(sample_intent, sample_entity, chatbot_features_df, intent_mapping)
print(f"Sample intent: {sample_intent}")
print(f"Sample entity: {sample_entity}")
print(f"Generated response: \"{response}\"")
```

```python
# 9. Complete Chatbot Function
print("\n\n== Integrating Chatbot Components ==")

def process_question(question, features_df, intent_mapping):
    """
    Process a user question and generate a response.
    
    Args:
        question: The user's question as a string
        features_df: DataFrame containing entity features
        intent_mapping: List of intent configurations
        
    Returns:
        A response string
    """
    # 1. Classify the intent
    intent = classify_intent(question)
    
    # 2. Extract the entity
    entity = extract_entity(question, features_df['entity'].tolist())
    
    print(f"Question: {question}")
    print(f"Classified Intent: {intent}")
    print(f"Extracted Entity: {entity}")
    
    # 3. If we couldn't determine the intent or entity, give a fallback response
    if not intent:
        return "I'm not sure I understand your question. Could you rephrase it? I can answer questions about anxiety prevalence, coping strategies, medication, therapy, social support, and lifestyle changes."
    
    if not entity:
        return "I need to know which country or region you're asking about. Could you specify?"
    
    # 4. Generate a response
    return generate_response(intent, entity, features_df, intent_mapping)

# Test the chatbot with sample questions
test_questions = [
    "What's the prevalence of anxiety in Australia?",
    "How do people in Germany cope with anxiety?",
    "Do people in Canada use medication for depression?",
    "Tell me about therapy options in United Kingdom",
    "What percentage of people in United States talk to friends about anxiety?",
    "How important is lifestyle change for people with anxiety in Sweden?",
    "What are the statistics for anxiety in Brazil?"
]

for question in test_questions:
    print("\n" + "-"*80)
    print(f"Q: {question}")
    response = process_question(question, chatbot_features_df, intent_mapping)
    print(f"A: {response}")
```

```python
# 10. Model Evaluation
print("\n\n== Model Evaluation ==")

# Create a test dataset for evaluation
def create_evaluation_dataset(features_df, intent_mapping):
    """Create a dataset for evaluating the chatbot"""
    # Sample entities for evaluation
    sample_entities = features_df['entity'].sample(min(10, len(features_df))).tolist()
    
    # Create test questions with expected intents and entities
    test_data = []
    
    # Template questions for each intent
    templates = {
        'anxiety_prevalence': [
            "What's the prevalence of anxiety in {entity}?",
            "How common is anxiety in {entity}?"
        ],
        'medication_info': [
            "Do people in {entity} use medication for anxiety?",
            "What percentage of people in {entity} take medication for depression?"
        ],
        'therapy_info': [
            "Do people in {entity} seek therapy for anxiety?",
            "What percentage of people in {entity} talk to professionals about mental health?"
        ],
        'coping_strategies': [
            "How do people in {entity} cope with anxiety?",
            "What are common ways to deal with depression in {entity}?"
        ],
        'social_support': [
            "Do people in {entity} talk to friends about anxiety?",
            "How important is family support for depression in {entity}?"
        ],
        'lifestyle_changes': [
            "How important are lifestyle changes for anxiety in {entity}?",
            "Do people in {entity} spend time outdoors to manage depression?"
        ]
    }
    
    # Generate test questions using templates and entities
    for intent, question_templates in templates.items():
        for entity in sample_entities:
            for template in question_templates:
                question = template.format(entity=entity)
                test_data.append({
                    'question': question,
                    'expected_intent': intent,
                    'expected_entity': entity
                })
    
    # Convert to DataFrame
    return pd.DataFrame(test_data)

# Evaluate the chatbot
eval_data = create_evaluation_dataset(chatbot_features_df, intent_mapping)
print(f"Created {len(eval_data)} test examples for evaluation")

# Process each test question
eval_data['predicted_intent'] = None
eval_data['predicted_entity'] = None
eval_data['response'] = None

for idx, row in eval_data.iterrows():
    # Classify intent
    eval_data.at[idx, 'predicted_intent'] = classify_intent(row['question'])
    
    # Extract entity
    eval_data.at[idx, 'predicted_entity'] = extract_entity(row['question'], chatbot_features_df['entity'].tolist())
    
    # Generate response
    if eval_data.at[idx, 'predicted_intent'] and eval_data.at[idx, 'predicted_entity']:
        response = generate_response(
            eval_data.at[idx, 'predicted_intent'], 
            eval_data.at[idx, 'predicted_entity'],
            chatbot_features_df,
            intent_mapping
        )
    else:
        response = "Could not generate response due to missing intent or entity."
    
    eval_data.at[idx, 'response'] = response
```

```python
# 11. Calculate Accuracy Metrics
print("\n\n== Calculating Accuracy Metrics ==")

# Calculate accuracy metrics
eval_data['intent_correct'] = eval_data['predicted_intent'] == eval_data['expected_intent']
eval_data['entity_correct'] = eval_data['predicted_entity'] == eval_data['expected_entity']
eval_data['fully_correct'] = eval_data['intent_correct'] & eval_data['entity_correct']

# Display evaluation results
intent_accuracy = eval_data['intent_correct'].mean() * 100
entity_accuracy = eval_data['entity_correct'].mean() * 100
combined_accuracy = eval_data['fully_correct'].mean() * 100

print("\nEvaluation Results:")
print(f"Intent Classification Accuracy: {intent_accuracy:.2f}%")
print(f"Entity Extraction Accuracy: {entity_accuracy:.2f}%")
print(f"Combined Accuracy: {combined_accuracy:.2f}%")

# Calculate per-intent metrics
intent_metrics = {}
for intent in eval_data['expected_intent'].unique():
    # Filter rows for this intent
    intent_rows = eval_data[eval_data['expected_intent'] == intent]
    correct = intent_rows['predicted_intent'] == intent
    
    # Calculate metrics
    accuracy = correct.mean() * 100
    count = len(intent_rows)
    
    intent_metrics[intent] = {
        'accuracy': accuracy,
        'count': count
    }

# Display intent-specific metrics
print("\nIntent-specific Accuracy:")
for intent, metrics in intent_metrics.items():
    print(f"- {intent.replace('_', ' ').title()}: {metrics['accuracy']:.2f}% ({metrics['count']} examples)")

# Display sample evaluation results
print("\nSample evaluation results:")
sample_results = eval_data.sample(min(5, len(eval_data)))
display(sample_results[['question', 'expected_intent', 'predicted_intent', 
                        'expected_entity', 'predicted_entity', 'intent_correct', 
                        'entity_correct', 'response']])
```

```python
# Let's use our enhanced intent classification in the main chatbot function
print("\n\n== Integrating Enhanced Intent Classification ==")

# First, let's create our enhanced training corpus
enhanced_intent_corpus = []
enhanced_intent_labels = []

# Create more diverse examples for each intent
for intent_data in intent_mapping:
    # Base templates for questions
    templates = [
        "What is the {keyword} of anxiety?",
        "Tell me about {keyword} for anxiety",
        "How many people {keyword} with anxiety?",
        "What's the {keyword} in different countries?",
        # Add more diverse phrasings
        "I want to know about {keyword} for depression",
        "Can you tell me the {keyword} stats?",
        "How does {keyword} work for mental health?",
        "Show me {keyword} information"
    ]
    
    # Create examples from keywords
    for keyword in intent_data["keywords"]:
        for template in templates:
            example = template.format(keyword=keyword)
            enhanced_intent_corpus.append(example)
            enhanced_intent_labels.append(intent_data["intent"])

print(f"Created enhanced training corpus with {len(enhanced_intent_corpus)} examples")
```

```python
# Initialize TF-IDF vectorizer with improved parameters
enhanced_vectorizer = TfidfVectorizer(
    ngram_range=(1, 3),  # Include 1, 2, and 3-grams
    min_df=1,
    max_df=0.9,
    sublinear_tf=True,  # Apply sublinear tf scaling
    stop_words='english'  # Remove English stopwords
)

# Create vectors for the enhanced training corpus
enhanced_intent_vectors = enhanced_vectorizer.fit_transform(enhanced_intent_corpus)

# Update our intent classification function
def classify_intent_enhanced(question):
    """Enhanced intent classification using better preprocessing and TF-IDF"""
    # Preprocess the question
    cleaned_question = question.lower()
    
    # Replace common variations
    cleaned_question = cleaned_question.replace("depression", "anxiety")
    cleaned_question = cleaned_question.replace("mental health", "anxiety")
    
    # Expand common abbreviations
    cleaned_question = cleaned_question.replace("stats", "statistics")
    cleaned_question = cleaned_question.replace("info", "information")
    cleaned_question = cleaned_question.replace("meds", "medication")
    
    # Transform the question
    question_vector = enhanced_vectorizer.transform([cleaned_question])
    
    # Calculate similarity to all training examples
    similarities = cosine_similarity(question_vector, enhanced_intent_vectors).flatten()
    
    # Get the best match and its score
    best_match_idx = np.argmax(similarities)
    best_score = similarities[best_match_idx]
    
    # Return the intent if the score is above threshold
    if best_score > 0.3:  # Higher threshold for better precision
        return enhanced_intent_labels[best_match_idx]
    else:
        return None
```

```python
# Update our main chatbot function
def process_question_enhanced(question, features_df, intent_mapping):
    """
    Process a user question with enhanced intent classification.
    
    Args:
        question: The user's question as a string
        features_df: DataFrame containing entity features
        intent_mapping: List of intent configurations
        
    Returns:
        A response string
    """
    # 1. Classify the intent with enhanced function
    intent = classify_intent_enhanced(question)
    
    # 2. Extract the entity
    entity = extract_entity(question, features_df['entity'].tolist())
    
    print(f"Question: {question}")
    print(f"Classified Intent (Enhanced): {intent}")
    print(f"Extracted Entity: {entity}")
    
    # 3. If we couldn't determine the intent or entity, give a fallback response
    if not intent:
        return "I'm not sure I understand your question. Could you rephrase it? I can answer questions about anxiety prevalence, coping strategies, medication, therapy, social support, and lifestyle changes."
    
    if not entity:
        return "I need to know which country or region you're asking about. Could you specify?"
    
    # 4. Generate a response
    return generate_response(intent, entity, features_df, intent_mapping)

# Test the enhanced chatbot with challenging examples
print("\n\n== Testing Enhanced Chatbot ==")

challenging_questions = [
    "How many people feel sad in Australia?",  # Different phrasing for prevalence
    "What methods help Germans manage their anxiety?",  # Different phrasing for coping
    "Do Canadians take pills for depression?",  # Different phrasing for medication
    "How many UK folks speak to professionals?",  # Different phrasing for therapy
    "Is talking to family common in the USA for handling stress?",  # Different phrasing for social support
    "Does being outdoors help Swedish people with mental health issues?"  # Different phrasing for lifestyle
]

# Test with challenging questions
for question in challenging_questions:
    print("\n" + "-"*80)
    print(f"Q: {question}")
    response = process_question_enhanced(question, chatbot_features_df, intent_mapping)
    print(f"A: {response}")
```

```python
# Generalized Mental Health Chatbot with Conversational Flow
import re

# Define a function to detect expressions of personal distress
def detect_personal_distress(message):
    """Detect if user is expressing feelings of depression or anxiety"""
    distress_phrases = [
        "i feel", "i am", "i'm", "feeling", "depressed", "anxious", 
        "anxiety", "depression", "sad", "down", "stressed", "overwhelmed",
        "worried", "unhappy", "miserable", "hopeless", "lost", "struggling"
    ]
    
    message_lower = message.lower()
    return any(phrase in message_lower for phrase in distress_phrases)

# Function to get a country/region from the message
def extract_location(message, entity_list):
    """Extract location from message or return None"""
    message_lower = message.lower()
    
    # Try to find entity in the message
    for entity in entity_list:
        if entity.lower() in message_lower:
            return entity
    
    return None

# Function to generate coping suggestions based on data
def generate_coping_suggestions(entity, features_df):
    """Generate personalized coping suggestions based on country data"""
    entity_data = features_df[features_df['entity'] == entity]
    
    if entity_data.empty:
        return "I don't have specific information for that country or region."
    
    entity_data = entity_data.iloc[0]
    
    # Get the dominant coping approach
    dominant_approach = entity_data.get('dominant_coping_approach', '')
    
    # Get the top 3 strategies
    strategies = [
        ('talked_to_friends_family', 'talking to friends or family'),
        ('spent_time_outdoors', 'spending time outdoors'),
        ('improved_lifestyle', 'making lifestyle improvements'),
        ('took_medication', 'taking medication'),
        ('talked_to_professional', 'talking to a mental health professional'),
        ('religious_spiritual_activities', 'engaging in religious or spiritual activities'),
        ('changed_relationships', 'making changes in relationships'),
        ('changed_work_situation', 'changing work situations')
    ]
    
    # Sort by value and get top 3
    sorted_strategies = sorted(
        [(strategy[0], strategy[1], entity_data.get(strategy[0], 0)) for strategy in strategies],
        key=lambda x: x[2],
        reverse=True
    )
    
    top_strategies = sorted_strategies[:3]
    
    # Create response
    response = f"In {entity}, many people manage their mental health by:\n\n"
    
    for i, (_, description, value) in enumerate(top_strategies):
        response += f"{i+1}. {description.capitalize()} ({value:.1f}% of people)\n"
    
    response += f"\nThe most common overall approach in {entity} is focused on {dominant_approach} strategies."
    
    # Add general disclaimer
    response += "\n\nRemember, if you're struggling with mental health issues, it's important to seek help from a qualified professional."
    
    return response

# Main conversation handling function
def process_user_message(message, features_df, conversation_state=None):
    """
    Process a user message in a conversational flow
    
    Args:
        message: User's message
        features_df: DataFrame with mental health features
        conversation_state: Optional dictionary to track conversation state
    
    Returns:
        response: Bot response
        updated_state: Updated conversation state
    """
    # Initialize or use existing state
    if conversation_state is None:
        conversation_state = {
            'awaiting_location': False,
            'distress_detected': False
        }
    
    # Extract entities/locations
    entity = extract_location(message, features_df['entity'].tolist())
    
    # Check if we're waiting for a location
    if conversation_state['awaiting_location']:
        if entity:
            # User provided location, generate coping suggestions
            response = generate_coping_suggestions(entity, features_df)
            conversation_state['awaiting_location'] = False
        else:
            # Still waiting for location
            response = "I need to know which country or region you're in to provide specific suggestions. Could you please tell me your country?"
        return response, conversation_state
    
    # Check if user is expressing distress
    if detect_personal_distress(message):
        conversation_state['distress_detected'] = True
        
        if entity:
            # User mentioned location, give suggestions
            response = generate_coping_suggestions(entity, features_df)
        else:
            # Need to ask for location
            response = "I'm sorry to hear you're feeling that way. To provide you with relevant suggestions, could you please tell me which country or region you're in?"
            conversation_state['awaiting_location'] = True
        
        return response, conversation_state
    
    # Otherwise, try to handle as a general question
    if entity:
        # Try to classify intent
        intent = classify_intent_enhanced(message)
        
        if intent:
            # Generate response based on intent and entity
            response = generate_response(intent, entity, features_df, intent_mapping)
        else:
            # Fall back to general information about the entity
            response = f"I have information about mental health in {entity}. You can ask about anxiety prevalence, coping strategies, medication use, therapy, social support, or lifestyle changes."
    else:
        # No entity and no distress detected - general response
        response = "I can provide information about mental health statistics and coping strategies around the world. If you're feeling anxious or depressed, I can suggest strategies based on your location. What country or region are you interested in?"
    
    return response, conversation_state

# Interactive test loop with conversation state tracking
def start_interactive_session(features_df):
    """Start an interactive session with conversation state tracking"""
    print("== Mental Health Chatbot Conversation ==")
    print("You can talk to me about mental health around the world.")
    print("If you're feeling anxious or depressed, I can suggest helpful strategies.")
    print("Type 'exit' to end the conversation.")
    print("\n")
    
    conversation_state = {
        'awaiting_location': False,
        'distress_detected': False
    }
    
    while True:
        user_input = input("You: ")
        
        if user_input.lower() == 'exit':
            print("Thank you for talking with me. Take care!")
            break
        
        response, conversation_state = process_user_message(
            user_input, 
            features_df, 
            conversation_state
        )
        
        print(f"Chatbot: {response}")
        print()

# To test this, run:
# start_interactive_session(chatbot_features_df)
```

```python
import pandas as pd
import numpy as np
import re
import random
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

# Enhanced chatbot response function with more natural language
def enhanced_chatbot_response(question, entity=None, print_question=True):
    """
    Generate a comprehensive response based on available data with natural language
    
    Args:
        question: The user's question
        entity: Optional entity (country/region) 
        print_question: Whether to print the question
        
    Returns:
        A response string
    """
    # Check for greetings first
    greeting_phrases = ["hello", "hi", "hey", "greetings", "good morning", "good afternoon", "good evening"]
    if question.lower() in greeting_phrases or any(phrase in question.lower() for phrase in greeting_phrases):
        greeting_responses = [
            "Hello! I'm a Mental Health Companion, here to provide information about mental health statistics and coping strategies around the world. How can I help you today?",
            "Hi there! I can provide information about mental health statistics and coping strategies globally. What would you like to know?",
            "Hello! I'm here to help with information about mental health across different countries. How can I assist you today?"
        ]
        return random.choice(greeting_responses)
    
    # Extract intent and entity if not provided
    intent = classify_intent_enhanced(question)
    if entity is None:
        entity = extract_entity(question, chatbot_features_df['entity'].tolist())
    
    # Only print question info if print_question is True
    if print_question:
        print(f"Question: {question}")
    
    # Check if the user is expressing personal distress
    if detect_personal_distress(question):
        if entity:
            # Use varied intros for emotional support
            distress_intros = [
                f"I understand you may be experiencing difficult emotions. In {entity}, many people manage their mental health by:",
                f"It sounds like you're going through a challenging time. Here's how people in {entity} commonly cope:",
                f"I'm sorry to hear you're feeling that way. In {entity}, these strategies are often helpful:"
            ]
            
            response = random.choice(distress_intros) + "\n\n"
            response += generate_coping_suggestions(entity, chatbot_features_df)
            return response
        else:
            location_requests = [
                "I notice you might be going through a difficult time. To provide you with the most relevant support information, could you tell me which country or region you're in?",
                "I'd like to offer some helpful strategies. Which country or region are you interested in?",
                "Different regions have different approaches to mental health support. Which country would you like information about?"
            ]
            return random.choice(location_requests)
    
    # Handle different intents
    if not intent or not entity:
        general_responses = [
            "I can provide information about mental health statistics and coping strategies around the world. Please ask about a specific country or region, such as 'How do people in Australia cope with anxiety?' or 'What's the prevalence of depression in Canada?'",
            "I have information about mental health in various countries. You can ask questions like 'How do people cope with anxiety in Germany?' or 'What's the rate of depression in Brazil?'",
            "To help you best, I need to know which country you're interested in. You can ask about coping strategies, prevalence rates, or support systems in specific regions."
        ]
        return random.choice(general_responses)
    
    # Generate response based on intent and entity
    if intent == "anxiety_prevalence":
        entity_data = chatbot_features_df[chatbot_features_df['entity'] == entity].iloc[0]
        prevalence = entity_data.get('anxiety_prevalence', 0)
        
        prevalence_responses = [
            f"In {entity}, approximately {prevalence:.1f}% of people reported experiencing anxiety or depression in 2020.",
            f"Based on our data, about {prevalence:.1f}% of people in {entity} report symptoms of anxiety or depression.",
            f"Mental health surveys in {entity} indicate that around {prevalence:.1f}% of the population experiences anxiety or depression."
        ]
        return random.choice(prevalence_responses)
    
    elif intent == "coping_strategies":
        return generate_coping_suggestions(entity, chatbot_features_df)
    
    else:
        # Use the regular response generation
        return generate_response(intent, entity, chatbot_features_df, intent_mapping)

# Detect personal distress in message
def detect_personal_distress(message):
    """Detect if a message indicates the user is experiencing distress"""
    distress_words = [
        "depressed", "anxious", "sad", "depression", "anxiety", "feel low", "feeling low", 
        "stressed", "stress", "worried", "upset", "unhappy", "miserable", "down",
        "hopeless", "helpless", "overwhelmed", "struggling", "can't cope", "difficult time"
    ]
    
    message_lower = message.lower()
    return any(word in message_lower for word in distress_words)

# Generate coping suggestions with more detailed information
def generate_coping_suggestions(entity, features_df):
    """Generate detailed coping suggestions based on data"""
    entity_data = features_df[features_df['entity'] == entity]
    
    if entity_data.empty:
        return f"I don't have specific information for {entity}."
    
    entity_data = entity_data.iloc[0]
    
    # Get the top strategies
    strategies = [
        ('talked_to_friends_family', 'Talking to friends or family'),
        ('spent_time_outdoors', 'Spending time outdoors'),
        ('improved_lifestyle', 'Making lifestyle improvements'),
        ('took_medication', 'Taking medication'),
        ('talked_to_professional', 'Talking to a mental health professional'),
        ('religious_spiritual_activities', 'Engaging in religious or spiritual activities')
    ]
    
    # Sort by value and get top 3
    sorted_strategies = sorted(
        [(key, desc, entity_data.get(key, 0)) for key, desc in strategies],
        key=lambda x: x[2],
        reverse=True
    )
    
    top_strategies = sorted_strategies[:3]
    
    # Create response
    response = ""
    for i, (_, description, value) in enumerate(top_strategies):
        response += f"{i+1}. {description} ({value:.1f}% of people)\n"
    
    # Add information about the dominant approach
    response += f"\nThe most common overall approach in {entity} is focused on {entity_data.get('dominant_coping_approach', 'various')} strategies."
    
    # Add professional help reminder
    response += "\n\nRemember, if you're struggling with mental health issues, it's important to seek help from a qualified professional."
    
    return response

# The complete naturalistic chatbot function
def interactive_chatbot():
    """Run a naturalistic interactive mental health chatbot"""
    print("== Mental Health Companion ==")
    print("Ask questions about mental health statistics and coping strategies around the world.")
    print("Type 'exit' to end the conversation.\n")
    
    # Enhanced conversation state with better context tracking
    context = {
        'current_entity': None,        # Most recent country mentioned
        'expressed_distress': False,   # Whether user expressed negative emotions
        'conversation_history': [],    # Track previous interactions
        'prev_responses': set(),       # Track previous responses to avoid repetition
        'question_count': 0            # Track number of questions asked
    }
    
    while True:
        user_input = input("You: ")
        
        if user_input.lower() == 'exit':
            print("Chatbot: Thank you for chatting with me today. Take care of yourself, and remember help is available if you need it.")
            break
        
        # Add to conversation history
        context['conversation_history'].append(user_input)
        context['question_count'] += 1
        
        # ----------------- GRATITUDE DETECTION -----------------
        gratitude_phrases = ["thank you", "thanks", "thank", "appreciate", "grateful", "thankyou"]
        if any(phrase in user_input.lower() for phrase in gratitude_phrases):
            gratitude_responses = [
                "You're welcome! I'm here to help. Is there anything else you'd like to know?",
                "Happy to help! Let me know if you have any other questions about mental health.",
                "No problem at all. Mental health is important, and I'm glad I could provide some information."
            ]
            # Select a response that hasn't been used yet if possible
            available_responses = [r for r in gratitude_responses if r not in context['prev_responses']]
            if not available_responses:
                available_responses = gratitude_responses
            
            response = random.choice(available_responses)
            context['prev_responses'].add(response)
            print(f"Chatbot: {response}\n")
            continue
        
        # ----------------- AFFIRMATIVE RESPONSES (YES/OK) -----------------
        affirmative_phrases = ["yes", "yeah", "sure", "ok", "okay", "yep", "yup"]
        if user_input.lower() in affirmative_phrases and context['question_count'] > 1:
            if context['current_entity']:
                entity = context['current_entity']
                response = f"Great! Here are some additional coping strategies that research has found effective in {entity}:\n\n"
                
                # Provide more specific advice
                response += " Regular physical exercise - Even a short walk can help reduce anxiety and improve mood\n"
                response += " Mindfulness and relaxation techniques - Simple breathing exercises or meditation\n"
                response += " Maintaining social connections - Regular contact with supportive people\n"
                response += " Limiting alcohol and caffeine - These can worsen anxiety and depression\n\n"
                
                response += "Would you like to know about specific mental health resources available in your area?"
            else:
                response = "I can provide information about mental health coping strategies, prevalence rates, and support options across different countries. Is there a specific country you're interested in learning about?"
            
            print(f"Chatbot: {response}\n")
            continue
        
        # ----------------- DETECT DISTRESS -----------------
        distress_words = ["depressed", "anxious", "sad", "depression", "anxiety", "feel low", "feeling low", 
                         "stressed", "stress", "worried", "upset", "unhappy", "miserable", "down"]
        
        newly_expressed_distress = any(word in user_input.lower() for word in distress_words)
        if newly_expressed_distress:
            context['expressed_distress'] = True
        
        # ----------------- ENTITY EXTRACTION -----------------
        # Extract entity from current message
        entity = extract_entity(user_input, chatbot_features_df['entity'].tolist())
        
        # If entity found, update context
        if entity:
            context['current_entity'] = entity
        
        # ----------------- GENERAL QUESTION HANDLING -----------------
        # Check for general prevalence questions
        prevalence_patterns = ["how many", "how common", "percentage", "prevalence", "statistics", "stats", "rate", "suffer from"]
        asking_general_prevalence = any(pattern in user_input.lower() for pattern in prevalence_patterns)
        
        if asking_general_prevalence and not entity and context['current_entity']:
            # They're asking a general question but we know their entity from context
            entity = context['current_entity']
            response = f"Based on our data for {entity}, anxiety and depression affect a significant portion of the population. "
            
            # Try to find prevalence data
            entity_data = chatbot_features_df[chatbot_features_df['entity'] == entity]
            if not entity_data.empty:
                prevalence = entity_data.iloc[0].get('anxiety_prevalence')
                if prevalence:
                    response += f"Approximately {prevalence:.1f}% of people report experiencing anxiety or depression. "
            
            response += "It's important to remember that many people don't report their symptoms, so actual rates may be higher."
            
            print(f"Chatbot: {response}\n")
            continue
        
        # Check for comparing countries
        compare_patterns = ["compare", "difference", "versus", "vs", "better", "worse", "than"]
        asking_comparison = any(pattern in user_input.lower() for pattern in compare_patterns)
        
        if asking_comparison and context['current_entity'] and entity and entity != context['current_entity']:
            # They want to compare two countries
            entity1 = context['current_entity']
            entity2 = entity
            
            # Get data for both entities
            entity1_data = chatbot_features_df[chatbot_features_df['entity'] == entity1].iloc[0]
            entity2_data = chatbot_features_df[chatbot_features_df['entity'] == entity2].iloc[0]
            
            response = f"Here's a comparison of mental health approaches between {entity1} and {entity2}:\n\n"
            
            # Compare prevalence
            prev1 = entity1_data.get('anxiety_prevalence', 0)
            prev2 = entity2_data.get('anxiety_prevalence', 0)
            response += f"Anxiety/Depression Prevalence:\n"
            response += f"- {entity1}: {prev1:.1f}%\n"
            response += f"- {entity2}: {prev2:.1f}%\n\n"
            
            # Compare top strategies
            response += f"Top Coping Strategy:\n"
            response += f"- {entity1}: {entity1_data.get('dominant_coping_approach', 'various')} approaches\n"
            response += f"- {entity2}: {entity2_data.get('dominant_coping_approach', 'various')} approaches\n"
            
            print(f"Chatbot: {response}\n")
            continue
        
        # ----------------- RESPONSE GENERATION -----------------
        # Normal processing with enhanced context awareness
        if context['expressed_distress'] and entity:
            # Personalized response for distress with known entity
            entity_response_key = f"distress_{entity}"
            if entity_response_key in context['prev_responses']:
                # Already gave coping strategies for this entity - provide different information
                response = f"I understand you're interested in mental health support in {entity}. "
                response += "In addition to the coping strategies I mentioned earlier, many people find it helpful to:\n\n"
                response += " Establish a consistent sleep schedule\n"
                response += " Break large tasks into smaller, manageable steps\n"
                response += " Practice self-compassion and be kind to yourself\n"
                response += " Limit exposure to negative news and social media\n\n"
                response += "Would you like information about professional mental health services?"
            else:
                # First time mentioning this entity with distress
                response = enhanced_chatbot_response(user_input, entity, print_question=True)
                context['prev_responses'].add(entity_response_key)
            
        elif context['expressed_distress'] and not entity:
            # They expressed distress but no entity - ask for location
            response = "I notice you might be experiencing difficult emotions. To provide you with relevant support information, which country or region are you interested in?"
            
        else:
            # Standard response for non-distress questions
            response = enhanced_chatbot_response(user_input, entity, print_question=True)
        
        print(f"Chatbot: {response}\n")

# Enhanced intent classification using TF-IDF
def classify_intent_enhanced(question):
    """Improved intent classification with more robust keyword detection"""
    # Map of high-level intent categories to specific keywords
    intent_keywords = {
        "anxiety_prevalence": ["prevalence", "common", "how many", "percentage", "rate", "statistics", 
                             "stats", "suffer", "affected", "have anxiety", "have depression", 
                             "diagnosed", "cases", "numbers", "data", "figures"],
        
        "medication_info": ["medication", "meds", "drugs", "prescriptions", "pills", "medicine", 
                         "pharmaceutical", "treatment", "antidepressants", "anti-anxiety", 
                         "medical treatment", "doctor prescribed"],
        
        "therapy_info": ["therapy", "counseling", "professional", "psychologist", "psychiatrist", 
                      "therapist", "mental health professional", "clinical", "treatment",
                      "psychological help", "counselor", "consultation"],
        
        "coping_strategies": ["cope", "deal", "manage", "handling", "strategies", "methods", "help",
                           "techniques", "approaches", "ways", "handle", "overcome", "combat",
                           "fight", "address", "confront", "face", "tackle"],
        
        "social_support": ["friends", "family", "talk", "social", "support", "relationships",
                        "loved ones", "peers", "community", "connection", "reaching out",
                        "network", "social circle"],
        
        "lifestyle_changes": ["lifestyle", "exercise", "diet", "outdoors", "nature", "physical", 
                           "habits", "routine", "activities", "hobbies", "nutrition", "sleep",
                           "workout", "fitness", "wellness", "healthy habits"]
    }
    
    # Lowercase the question for case-insensitive matching
    question_lower = question.lower()
    
    # Score each intent based on keyword matches
    intent_scores = {}
    for intent, keywords in intent_keywords.items():
        score = 0
        for keyword in keywords:
            if keyword in question_lower:
                # Exact match gets higher score
                score += 1
            elif any(word in keyword for word in question_lower.split()):
                # Partial match gets lower score
                score += 0.5
        intent_scores[intent] = score
    
    # Get the highest scoring intent
    max_score = max(intent_scores.values())
    if max_score > 0:
        # Find all intents with the max score
        top_intents = [intent for intent, score in intent_scores.items() if score == max_score]
        return top_intents[0]  # Return the first if there are ties
    
    return None
```

```python
# Function to load, process, and save the depression disorder data
def process_and_save_depression_data():
    try:
        # Adjust the path to match where your file is located
        input_path = 'IDS/cap5771sp25-project/Data/Mental health Depression disorder Data.csv'
        print(f"Loading depression data from: {input_path}")
        
        depression_df = pd.read_csv(input_path)
        print(f"Successfully loaded depression disorder data with shape: {depression_df.shape}")
        
        # Basic cleaning and processing
        # Convert percentage columns to numeric if needed
        numeric_columns = depression_df.columns[depression_df.columns.str.contains('%')]
        for col in numeric_columns:
            depression_df[col] = pd.to_numeric(depression_df[col], errors='coerce')
        
        # Handle missing values
        depression_df = depression_df.fillna(0)  # Replace NaNs with 0
        
        # Create output directory if it doesn't exist
        output_dir = 'IDS/cap5771sp25-project/processed_data'
        os.makedirs(output_dir, exist_ok=True)
        
        # Save the processed data
        output_path = os.path.join(output_dir, 'processed_depression_data.csv')
        depression_df.to_csv(output_path, index=False)
        
        print(f"Processed depression data saved to: {output_path}")
        return depression_df
    
    except Exception as e:
        print(f"Error processing depression data: {e}")
        print(f"Current working directory: {os.getcwd()}")
        return None
```

```python
def simple_mental_health_chatbot():
    """Simple mental health chatbot following a specific conversational flow"""
    print("== Mental Health Assistant ==")
    print("Type 'exit' to end the conversation.\n")
    
    # Load the data
    try:
        coping_df, depression_df, disorders_df = load_processed_data()
        
        # Fallback if data loading fails
        if coping_df is None:
            print("Warning: Using fallback data as primary data couldn't be loaded")
            # Create a simple fallback dataset
            coping_df = pd.DataFrame({
                'entity': ['Australia', 'Canada', 'United States', 'United Kingdom', 'India', 'Germany'],
                'anxiety_prevalence': [20.1, 19.5, 23.4, 22.1, 15.8, 21.9],
                'dominant_coping_approach': ['lifestyle', 'lifestyle', 'clinical', 'social', 'spiritual', 'lifestyle'],
                'talked_to_friends_family': [83.7, 84.2, 80.1, 79.6, 82.8, 87.4],
                'improved_lifestyle': [78.5, 83.8, 76.5, 72.3, 77.6, 82.5],
                'spent_time_outdoors': [79.7, 82.2, 75.4, 77.2, 71.7, 81.2],
                'took_medication': [71.4, 73.0, 78.6, 69.1, 63.4, 65.1],
                'talked_to_professional': [75.4, 69.5, 71.2, 84.3, 68.2, 70.5],
                'religious_spiritual_activities': [19.4, 21.3, 30.2, 18.5, 60.3, 15.2]
            })
    except Exception as e:
        print(f"Error loading data: {e}")
        # Create a simple fallback dataset
        coping_df = pd.DataFrame({
            'entity': ['Australia', 'Canada', 'United States', 'United Kingdom', 'India', 'Germany'],
            'anxiety_prevalence': [20.1, 19.5, 23.4, 22.1, 15.8, 21.9],
            'dominant_coping_approach': ['lifestyle', 'lifestyle', 'clinical', 'social', 'spiritual', 'lifestyle'],
            'talked_to_friends_family': [83.7, 84.2, 80.1, 79.6, 82.8, 87.4],
            'improved_lifestyle': [78.5, 83.8, 76.5, 72.3, 77.6, 82.5],
            'spent_time_outdoors': [79.7, 82.2, 75.4, 77.2, 71.7, 81.2],
            'took_medication': [71.4, 73.0, 78.6, 69.1, 63.4, 65.1],
            'talked_to_professional': [75.4, 69.5, 71.2, 84.3, 68.2, 70.5],
            'religious_spiritual_activities': [19.4, 21.3, 30.2, 18.5, 60.3, 15.2]
        })
        depression_df = None
        disorders_df = None
    
    # Welcome message
    print("Chatbot: Hi, welcome to mental health assistance. How are you today?\n")
    
    # Rest of your chatbot code remains the same...
    
    # Context tracking
    context = {
        'country': None,
        'days_feeling_low': None,
        'specific_worry': None,
        'disorder_type': None,
        'substance_use': None
    }
    
    # Conversation state
    state = "initial_greeting"
    
    while True:
        user_input = input("You: ")
        
        if user_input.lower() == 'exit':
            print("Chatbot: Thank you for talking with me. Take care and remember to seek professional help if needed.")
            break
        
        # Handle different conversation states
        if state == "initial_greeting":
            # Check if user is expressing negative feelings
            negative_feelings = ["bad", "low", "sad", "depressed", "anxious", "stressed", "unhappy"]
            
            if any(word in user_input.lower() for word in negative_feelings):
                print("Chatbot: From how many days you are feeling this?\n")
                state = "ask_duration"
            else:
                print("Chatbot: I'm here to help with mental health concerns. Are you experiencing any difficulties with your mood or emotions?\n")
                state = "check_wellbeing"
        
        elif state == "check_wellbeing":
            if "yes" in user_input.lower() or any(word in user_input.lower() for word in ["bad", "low", "sad", "depressed", "anxious"]):
                print("Chatbot: From how many days you are feeling this?\n")
                state = "ask_duration"
            else:
                print("Chatbot: That's good to hear. I'm here if you ever need to talk about mental health. Which country are you from?\n")
                state = "ask_country"
        
        elif state == "ask_duration":
            # Store the duration
            context['days_feeling_low'] = user_input
            
            # Move to asking about specific worry
            print("Chatbot: Are you worried about something specifically?\n")
            state = "ask_specific_worry"
        
        elif state == "ask_specific_worry":
            # Store the specific worry information
            context['specific_worry'] = user_input
            
            # Try to determine disorder type based on keywords
            if any(word in user_input.lower() for word in ["anxious", "anxiety", "worry", "panic", "fear"]):
                context['disorder_type'] = "anxiety"
            elif any(word in user_input.lower() for word in ["sad", "depressed", "hopeless", "unmotivated", "tired"]):
                context['disorder_type'] = "depression"
            else:
                context['disorder_type'] = "general"
            
            # Ask about substance use
            print("Chatbot: Do you have any alcohol or drug use habits?\n")
            state = "ask_substance_use"
        
        elif state == "ask_substance_use":
            # Store substance use information
            if any(word in user_input.lower() for word in ["yes", "sometimes", "occasionally", "often"]):
                context['substance_use'] = "yes"
            else:
                context['substance_use'] = "no"
            
            # Ask about country
            print("Chatbot: Which country you live in?\n")
            state = "ask_country"
        
        elif state == "ask_country":
            # Try to find the country in our database
            country = None
            
            # First look in coping strategies data
            for possible_country in coping_df['entity'].unique():
                if user_input.lower() in possible_country.lower() or possible_country.lower() in user_input.lower():
                    country = possible_country
                    break
            
            # If not found and we have depression data, look there
            if not country and depression_df is not None:
                for possible_country in depression_df['Entity'].unique():
                    if user_input.lower() in possible_country.lower() or possible_country.lower() in user_input.lower():
                        country = possible_country
                        break
            
            if country:
                context['country'] = country
                
                # Find disorder prevalence
                prevalence = None
                disorder_name = "depression" if context['disorder_type'] == "depression" else "anxiety"
                
                # Try to get depression specific data
                if depression_df is not None and context['disorder_type'] == "depression":
                    country_depression_data = depression_df[depression_df['Entity'] == country]
                    if not country_depression_data.empty and 'Depression (%)' in country_depression_data.columns:
                        prevalence = country_depression_data['Depression (%)'].values[0]
                
                # Fall back to anxiety data if we don't have depression data
                if prevalence is None and 'anxiety_prevalence' in coping_df.columns:
                    country_coping_data = coping_df[coping_df['entity'] == country]
                    if not country_coping_data.empty:
                        prevalence = country_coping_data['anxiety_prevalence'].values[0]
                
                # Default if no data found
                if prevalence is None:
                    prevalence = 10.0  # Fallback value
                
                # Get dominant approach
                dominant_approach = "various"
                country_coping_data = coping_df[coping_df['entity'] == country]
                if not country_coping_data.empty and 'dominant_coping_approach' in country_coping_data.columns:
                    dominant_approach = country_coping_data['dominant_coping_approach'].values[0]
                
                # Formulate response
                print(f"Chatbot: In {country}, about {prevalence:.1f}% of people experience symptoms of {disorder_name}. I can share some common coping strategies that people in your country find helpful. Would you like to see these strategies?\n")
                state = "offer_coping_strategies"
            else:
                print("Chatbot: I don't have data for that country. Could you try another country like Australia, Canada, United States, or India?\n")
        
        elif state == "offer_coping_strategies":
            if "yes" in user_input.lower():
                # They want coping strategies
                if context['country']:
                    # Find country data
                    country_data = coping_df[coping_df['entity'] == context['country']]
                    
                    if not country_data.empty:
                        country_data = country_data.iloc[0]
                        
                        # Add strategies based on their disorder and substance use
                        response = ""
                        
                        # Define strategies to check
                        strategy_keys = [
                            'talked_to_friends_family',
                            'improved_lifestyle',
                            'spent_time_outdoors',
                            'took_medication',
                            'talked_to_professional',
                            'religious_spiritual_activities',
                            'changed_work_situation',
                            'changed_relationships'
                        ]
                        
                        strategy_labels = [
                            'Talking to friends or family',
                            'Making lifestyle improvements',
                            'Spending time outdoors',
                            'Taking medication',
                            'Talking to a mental health professional',
                            'Engaging in religious or spiritual activities',
                            'Changing work situation',
                            'Changing relationships'
                        ]
                        
                        # Build list of strategies with values
                        strategies = []
                        for i, key in enumerate(strategy_keys):
                            if key in country_data and not pd.isna(country_data[key]):
                                strategies.append((strategy_labels[i], country_data[key]))
                        
                        # Sort by value (descending)
                        strategies.sort(key=lambda x: x[1], reverse=True)
                        
                        # Show top 3
                        for i, (label, value) in enumerate(strategies[:3]):
                            response += f"{i+1}. {label}\n"
                        
                        # Add specific note for substance users
                        if context['substance_use'] == "yes":
                            response += "\nSince you mentioned substance use, it's important to know that reducing alcohol and drug use can significantly improve mental health symptoms.\n"
                        
                        dominant_approach = country_data.get('dominant_coping_approach', 'various')
                        response += f"\nThe most common overall approach in {context['country']} is focused on {dominant_approach} strategies."
                        print(f"Chatbot: {response}\n")
                    else:
                        print(f"Chatbot: I don't have detailed coping strategy data for {context['country']}, but generally, talking to friends/family, improving lifestyle habits, and seeking professional help are effective approaches for mental health concerns.\n")
                    
                    state = "await_feedback"
                else:
                    print("Chatbot: I need to know which country you're from to provide relevant strategies. Which country do you live in?\n")
                    state = "ask_country"
            else:
                print("Chatbot: That's okay. Is there something else you'd like to talk about regarding mental health?\n")
                state = "general_conversation"
        
        elif state == "await_feedback" or state == "general_conversation":
            if any(word in user_input.lower() for word in ["thanks", "thank you", "thankyou", "thank", "ok", "okay"]):
                print("Chatbot: You're welcome. If anything serious I suggest to see a professional doctor. If anything needed let me know.\n")
                state = "general_conversation"
            else:
                print("Chatbot: Is there anything else I can help you with regarding mental health?\n")
                state = "general_conversation"

# Run the chatbot
simple_mental_health_chatbot()
```
